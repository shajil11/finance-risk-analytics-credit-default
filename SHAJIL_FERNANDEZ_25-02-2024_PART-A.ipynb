### 1.1) Outlier treatment

```python
# creating boxplot
df.boxplot(figsize=(15,10))
plt.xticks(rotation=90)
plt.show()
```

```python
Default_X = df.drop(['Default', 'Co_Code', '_Liability_Assets_Flag', 'Co_Name'], axis = 1)
Default_Y = df[['Co_Code', 'Co_Name', '_Liability_Assets_Flag', 'Default']]
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
for column in Default_X.columns:
    lr,ur=remove_outlier(df[column])
    Default_X[column]=np.where(Default_X[column]>ur,ur,Default_X[column])
    Default_X[column]=np.where(Default_X[column]<lr,lr,Default_X[column])
```

```python
df_sub1 = pd.concat([Default_X, Default_Y], axis = 1)
```

```python
df_sub1.boxplot(figsize=(15,10))
plt.xticks(rotation=90)
plt.show()
```

```python
# VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```
